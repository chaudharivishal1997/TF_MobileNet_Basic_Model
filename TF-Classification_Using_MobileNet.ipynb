# MobileNet

```python
# Importing required libraries
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Model
from tensorflow.keras.applications import imagenet_utils
from sklearn.metrics import confusion_matrix
import itertools
import shutil
import os
import random
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Downloading MobileNet model
mobile = tf.keras.applications.mobilenet.MobileNet()
```

```python
# Defining function to preprocess images
def prepare_image(file):
    img_path = 'C:/Users/vishal.chaudhari/MobileNet-samples/'
    img = image.load_img(img_path+file, target_size=(224,224))
    img_array = image.img_to_array(img)
    img_array_expanded_dims = np.expand_dims(img_array, axis=0)
    return tf.keras.applications.mobilenet.preprocess_input(img_array_expanded_dims)
```

```python
# Plotting random images (we've just created random images dataset to use)
from IPython.display import Image

Image(filename='C:/Users/vishal.chaudhari/MobileNet-samples/1.jpg', width=300, height=200)
```

```python
# Making predictions using MobileNet
preprocessed_image = prepare_image('1.jpg')
predictions = mobile.predict(preprocessed_image)
results = imagenet_utils.decode_predictions(predictions) # decoding top 5 predictions from ImageNet 1000 categories
results
```

```python
# Persian_cat have highest probability
assert results[0][0][1] == 'Persian_cat'  
```
